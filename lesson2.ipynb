Figure и Axes - основные объекты библиотеки Matplotlib
Объекты Axes распологаются внутри figure

```python
x = np.linspace(-5, 5, 101)
x2 = x**2
x3 = x **3
```

```python
plt.axis([-5, 5, -8, 8])
plt.plot(x, x2)
plt.plot(x, x3)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(x, x2)
```

Объекты figure axes

```python
# Создание объектов
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax
ax1.plot(x, x2)
ax2.plot(x, x3)
```

```python
# Метод ax.flatten()
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
```

```python
# set_size_inches() - меняем размер графиков
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
fig.set_size_inches(8, 3.5)
```

```python
# Увеличиваем расстояние между графиками 
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
fig.set_size_inches(8, 3.5)

plt.subplots_adjust(wspace=0.3)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1)
print(ax.shape)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
fig.set_size_inches(3, 5)
```

```python
#plt.subplots_adjust(hspace=0.25) - расстояние между графиками

fig, ax = plt.subplots(nrows=2, ncols=1)
print(ax.shape)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
fig.set_size_inches(3, 5)
plt.subplots_adjust(hspace=0.25)
```

```python
y = np.sin(x)
z = np.cos(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)

ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, x2)
ax1.set_title('График x**2')

ax2.plot(x, x3)
ax2.set_title('График x**3')

ax3.plot(x, y)
ax3.set_title('График sin(x)')

ax4.plot(x, z)
ax4.set_title('График sin(x)')

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

```python
#sharey=True

fig, ax = plt.subplots(nrows=2, ncols=2, sharey=True)

ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, x2)
ax1.set_title('График x**2')

ax2.plot(x, x3)
ax2.set_title('График x**3')

ax3.plot(x, y)
ax3.set_title('График sin(x)')

ax4.plot(x, z)
ax4.set_title('График sin(x)')

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

```python
#устанавливаем границы для каждого графика ax1.set_xlim

fig, ax = plt.subplots(nrows=2, ncols=2, sharey=True)

ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, x2)
ax1.set_title('График x**2')
ax1.set_xlim([-3, 3])

ax2.plot(x, x3)
ax2.set_title('График x**3')

ax3.plot(x, y)
ax3.set_title('График sin(x)')

ax4.plot(x, z)
ax4.set_title('График sin(x)')

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

```python
#устанавливаем границы для каждого графика ax2.set_ylim

fig, ax = plt.subplots(nrows=2, ncols=2, sharey=True)

ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, x2)
ax1.set_title('График x**2')
ax1.set_xlim([-3, 3])

ax2.plot(x, x3)
ax2.set_title('График x**3')
ax2.set_ylim([-25, 25])


ax3.plot(x, y)
ax3.set_title('График sin(x)')

ax4.plot(x, z)
ax4.set_title('График sin(x)')

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```
